## Libraries

```python
import pandas as pd
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
Xco2_L3_df = pd.read_csv('csv_files/L3/Xco2_L3_2015_to_22.csv')
```

```python
Xco2_L3_df
```

## XCo2 Distribution

```python
#Check the distribution of values
Xco2_L3_df.groupby('Year')['Month'].value_counts().unstack().T
```

All 12 month entries are well put + the 2 months in 2022

```python
print("MAX: ", np.max(Xco2_L3_df['xco2']))
print("MIN: ", np.min(Xco2_L3_df['xco2']))
```

```python
# Plot Xco2 Distribution

plt.figure(figsize = (12,6))
sns.displot(Xco2_L3_df.xco2)
plt.title('Averaged Xco2 ppm distribution from 2015 to 2022')
plt.set_ylabel = ('count')
plt.set_xlabel = ('Xco2 ppm')

plt.show();
```

## Trend

```python
avg_xco2 = Xco2_L3_df[['xco2', 'Year' ,'Month']].groupby(['Year','Month']).mean().unstack()
avg_xco2
```

```python
avg_xco2['xco2']['May']
```

```python
plt.title("May's Average xco2 ppm", fontsize = 20)
sns.lineplot(data=avg_xco2['xco2']['May'])
plt.ylabel("xco2 ppm", fontsize = 16)
plt.xlabel("Year", fontsize = 16)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16);
```

```python
plt.title("Average xco2 ppm", fontsize = 20)
sns.lineplot(data=avg_xco2['xco2'])
plt.ylabel("xco2 ppm", fontsize = 16)
plt.xlabel("Year", fontsize = 16)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16);
```

```python
Northern_hemisphere = Xco2_L3_df[Xco2_L3_df['latitude'] >= 0]
Northern_hemisphere
```

```python
avg_xco2_North = Northern_hemisphere[['xco2', 'Year' ,'Month']].groupby(['Year','Month']).mean().unstack()
avg_xco2_North
```

```python
Southern_hemisphere = Xco2_L3_df[Xco2_L3_df['latitude'] < 0]
Southern_hemisphere
```

```python
avg_xco2_South = Southern_hemisphere[['xco2', 'Year' ,'Month']].groupby(['Year','Month']).mean().unstack()
avg_xco2_South
```

```python
# Plot Northern v/s Southern Hemisphere
fig, ax = plt.subplots(2,1, figsize = (16,16))
g1 = sns.lineplot(data=avg_xco2_North['xco2'], ax = ax[0])
g1.set_title("Northern Hemisphere Average xco2 ppm", fontsize = 20)
g1.set_ylabel("xco2 ppm", fontsize = 16)
g1.set_xlabel("Year", fontsize = 16)
g1.tick_params(axis = 'x', labelsize = 16)
g1.tick_params(axis = 'y', labelsize = 16)


g2 = sns.lineplot(data=avg_xco2_South['xco2'], ax = ax[1])
g2.set_title("Southern Hemisphere Average xco2 ppm", fontsize = 20)
g2.set_ylabel("xco2 ppm", fontsize = 16)
g2.set_xlabel("Year", fontsize = 16)

plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16);
```

```python
Northern_avg_xco2 = Northern_hemisphere[['xco2', 'Year' ,'Month']].groupby(['Year','Month']).mean().reset_index()
Northern_avg_xco2
```

```python
Southern_avg_xco2 = Southern_hemisphere[['xco2', 'Year' ,'Month']].groupby(['Year','Month']).mean().reset_index()
Southern_avg_xco2
```

```python
#Sort by month
month_dict = {'January':1,'February':2,'March':3, 'April':4, 'May':5, 'June':6, 'July':7, 'August':8, 'September':9, 
              'October':10, 'November':11, 'December':12}
```

```python
Northern_avg_xco2.sort_values('Month', key = lambda x : x.apply (lambda x : month_dict[x]), inplace = True)
Northern_avg_xco2
```

```python
Southern_avg_xco2.sort_values('Month', key = lambda x : x.apply (lambda x : month_dict[x]), inplace = True)
Southern_avg_xco2
```

```python
Jan_Xco2_North = Northern_avg_xco2.query("Month == 'January'")
Jan_Xco2_South = Southern_avg_xco2.query("Month == 'January'")
sns.lineplot(data = Jan_Xco2_North, x = 'Year', y = 'xco2', label = 'Northern Hemisphere')
sns.lineplot(data = Jan_Xco2_South, x = 'Year', y = 'xco2', label = 'Southern Hemisphere')
plt.title("January's Average xco2 ppm", fontsize =20)
plt.xlabel('Year', fontsize = 16)
plt.ylabel('xco2', fontsize = 16)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.legend(fontsize = 'large');
```

## Hemispheric Seasonal Cycles 
Global CO2 concentrations fluctuate seasonally with CO2 absorption and respiration by vegetation over the annual cycle. Seasonal fluctuations are more pronounced in the Northern Hemisphere because it has more plant life

```python
# Plot Avg xco2 with 95% confidence interval
fig, ax = plt.subplots(2,1, figsize = (16,16))
plt.suptitle('2015 -  2022 Average xco2 ppm with 95% Confidence Interval', fontsize = 20)
g1 = sns.lineplot(data = Northern_avg_xco2, x = 'Month', y = 'xco2', ax = ax[0])
g1.set_title("Northern Hemisphere", fontsize = 18)
g1.set_ylabel("xco2 ppm", fontsize = 16)
g1.set_xlabel("Year", fontsize = 16)
g1.tick_params(axis = 'x', labelsize = 16, labelrotation = 0)
g1.tick_params(axis = 'y', labelsize = 16)
# g1.set_xticklabels(g1.get_xticklabels(),rotation=90)

g2 = sns.lineplot(data = Southern_avg_xco2, x = 'Month', y = 'xco2', ax = ax[1])
g2.set_title("Southern Hemisphere", fontsize = 18)
g2.set_ylabel("xco2 ppm", fontsize = 16)
g2.set_xlabel("Year", fontsize = 16)

plt.xticks(fontsize = 16, rotation = 0)
plt.yticks(fontsize = 16);
```

```python
plt.figure(figsize = (16, 6))
plt.title("2015 - 2020 Average xco2 ppm ", fontsize = 20)
sns.lineplot(data = Northern_hemisphere[['xco2', 'Month']].groupby(['Month']).mean().
             sort_values('Month', key = lambda x : x.apply (lambda x : month_dict[x])), x = 'Month', y = 'xco2',
            label = 'Northern Hemisphere')
sns.lineplot(data = Southern_hemisphere[['xco2', 'Month']].groupby(['Month']).mean().
             sort_values('Month', key = lambda x : x.apply (lambda x : month_dict[x])), x = 'Month', y = 'xco2',
            label = 'Southern Hemisphere')
plt.legend(fontsize = 'large')
plt.ylabel('xco2 ppm', fontsize = 16)
plt.xlabel('Year', fontsize = 16)
plt.xticks(fontsize = 16, rotation = 0)
plt.yticks(fontsize = 16);
```
